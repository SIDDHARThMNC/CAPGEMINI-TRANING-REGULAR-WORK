```python
# Scenario: AI System for Detecting Fraud in Financial Transactions
# The Problem
# A major bank processes millions of transactions daily.
# Fraud analysts are overwhelmed by the sheer volume and must make instant decisions.
# Critical cases like credit card fraud, money laundering must be flagged in real-time.

# The Solution: Build an AI Assistant
# The bank deploys an AI model that can pre-screen transactions and alert analysts.
# Enter Transfer Learning:
# Instead of starting from zero, engineers use BERT trained on massive text corpora.
# By fine-tuning BERT on the bank's fraud dataset, the AI can quickly learn to distinguish:
# - Suspicious vs. normal transactions
# - Fraud rings vs. genuine customer activity
```

```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from datasets import Dataset
import torch

# Sample fraud dataset
data = {
    "text": [
        "Purchase at local grocery store",
        "International wire transfer to unknown account",
        "ATM withdrawal at home city",
        "Multiple rapid transactions in different countries",
        "Online shopping at known retailer",
        "Large cash deposit followed by immediate withdrawal"
    ],
    "label": [0, 1, 0, 1, 0, 1]  # 0 = normal, 1 = fraud
}

df = pd.DataFrame(data)

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Tokenize
def tokenize(batch):
    return tokenizer(batch["text"], padding=True, truncation=True, max_length=64)

dataset = Dataset.from_pandas(df)
dataset = dataset.map(tokenize, batched=True)

# Load model
model = AutoModelForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2)

print("Model ready for fine-tuning")
print(df)
```
