```python
# Project Scenario 2: AI Research Assistant for Students
# A university wants to build an AI-powered Research Assistant that can help students quickly understand long academic articles and generate structured insights.

# The system should be able to:
# - Summarize long text
# - Extract key insights
# - Answer questions about the content
# - Generate concise explanations

# Task 1: LLM Interaction Setup
# Task 2: Prompt Engineering Experiments (Zero-Shot, Few-Shot, Chain-of-Thought)
# Task 3: Prompt Optimization
# Task 4: Tokenization Experiment
# Task 5: Build a Mini AI Tool
```

```python
# Task 1: LLM Interaction Setup
from langchain.chat_models import init_chat_model
from dotenv import load_dotenv

load_dotenv()

model = init_chat_model("mistral-medium-latest")

article = """
Artificial intelligence (AI) is rapidly transforming industries worldwide.
From healthcare to finance, AI systems are being deployed to automate tasks,
improve decision-making, and enhance customer experiences.
Machine learning, a subset of AI, enables systems to learn from data without being explicitly programmed.
Deep learning, using neural networks, has achieved remarkable results in image recognition, natural language processing, and game playing.
"""

summary_prompt = f"Summarize the following article in 3 bullet points:\n\n{article}"
response = model.invoke(summary_prompt)
print("Summary:")
print(response.content)
```

```python
# Task 2: Prompt Engineering Experiments

# Zero-Shot Prompt
zero_shot = f"Summarize the following article in 5 bullet points.\n\n{article}"
print("Zero-Shot Output:")
print(model.invoke(zero_shot).content)

# Few-Shot Prompt
few_shot = f"""
Example 1:
Article: The sun is a star at the center of our solar system.
Summary: The sun is the central star of our solar system.

Example 2:
Article: Python is a high-level programming language known for its simplicity.
Summary: Python is a simple, high-level programming language.

Now summarize this article:
{article}
"""
print("\nFew-Shot Output:")
print(model.invoke(few_shot).content)

# Chain-of-Thought Prompt
cot = f"""
Analyze the article step by step:
1. Identify the main topic
2. Extract key ideas
3. Generate a concise summary

Article:
{article}
"""
print("\nChain-of-Thought Output:")
print(model.invoke(cot).content)
```

```python
# Task 3: Prompt Optimization
optimized_prompt = f"""
You are an expert research analyst. Read the following article carefully and provide:
1. Three key insights
2. One actionable takeaway
Use a professional tone.

Article:
{article}
"""
print("Optimized Output:")
print(model.invoke(optimized_prompt).content)
```

```python
# Task 4: Tokenization Experiment
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

tokens = tokenizer.tokenize(article)
print(f"Total tokens: {len(tokens)}")
print(f"Tokens: {tokens[:20]}...")
```

```python
# Task 5: Mini AI Tool
def research_assistant(text):
    prompt = f"""
    You are a research assistant. Given the following article:
    
    {text}
    
    Provide:
    1. A short summary (2-3 sentences)
    2. Key insights (3 bullet points)
    3. One actionable recommendation
    """
    response = model.invoke(prompt)
    return response.content

user_input = input("Paste your article here: ")
print(research_assistant(user_input))
```
